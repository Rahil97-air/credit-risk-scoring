## 3. Key Findings

### 3.1 Class Imbalance
Only ~8% of loans end in default. This severe imbalance requires:
- SMOTE oversampling in training
- `class_weight='balanced'` in all classifiers
- F2-score (not accuracy) as the optimisation metric

### 3.2 Missing Values
Many columns have >40% missing (e.g. `EXT_SOURCE_1`, `OWN_CAR_AGE`).  
These are dropped in preprocessing if >50% missing; otherwise median-imputed.

### 3.3 External Scores
`EXT_SOURCE_2` is the single strongest predictor (highest negative correlation with TARGET).  
Lower external scores → higher default probability.

### 3.4 Income & Default
Lowest income quintile shows ~12% default rate vs ~6% for the highest — a 2× difference.  
This confirms income-based segmentation is meaningful.

### 3.5 Education Type
Applicants with 'Lower secondary' education default at the highest rate (~11%).  
Those with 'Academic degree' show <5% — a 2× gap to monitor for fairness.

### 3.6 Contract Type
'Revolving loans' have meaningfully higher default rates than 'Cash loans'.
